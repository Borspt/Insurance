```python
fig, axes = plt.subplots(3, 4, figsize=(20, 30), sharey=True)
ax = axes.ravel()
fig.suptitle('Распределение категориальных признаков',
             y = 0.92,
             fontsize = 18, fontweight='bold')

sub_cat = [cat_cols[1], cat_cols[3]]
sub_cat.extend(cat_cols[6:])
                                            
for i, cat_col in enumerate(sub_cat):
    val1 = data_1[cat_col].value_counts()
    val0 = data_0[cat_col].value_counts()

    bars0 = ax[i].bar(height=val0,
                    x=range(len(val0.index)),
                    color='royalblue',
                    alpha=0.8,
                    label='Не пролонгирован')
                    
    
    
    bars1 = ax[i].bar(height=val1,
                    x=range(len(val1.index)),
                    color='darkgray',
                    label='Пролонгирован')
    
    if cat_col in ['VIP', 'deduct', 'kvs_type']:
        ax[i].set_xticks(range(len(val0.index)))  
        ax[i].set_xticklabels(list(val0.index), rotation=-15, ha='left')
    else:
        ax[i].set_xticks(range(len(val0.index)))  
        ax[i].set_xticklabels(list(val0.index))
    
    for bar in bars1:
        yval = bar.get_height()
        ax[i].text(bar.get_x()+0.4*bar.get_width(),
                   yval * 1.01,
                   yval,
                   color='black',
                   fontweight='bold')

    for bar in bars0:
        yval = bar.get_height()
        ax[i].text(bar.get_x(),
                   yval * 1.01,
                   yval,
                   color='darkblue',
                   fontweight='bold')

        ax[i].set_title(cat_col)
        ax[i].legend(loc='upper right')

        


plt.show()
```
